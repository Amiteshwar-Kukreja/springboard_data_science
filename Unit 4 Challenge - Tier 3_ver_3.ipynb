```python
# The modelling logic below has 5 steps as enumerated in the comments


# Step 1: Create a pivot table that returns an annual Average_Price for each Borough. The Average_Price is the mean 
#         of the 12 monthly observations for each year.

avg_annual_price = properties_tall.pivot_table(values="Average_Price",index="London_Borough",columns="year")
avg_annual_price = avg_annual_price.reset_index()

# Step 2: Keep data for the year 2000 to 2020. Drop the remanining years. 

avg_annual_price = avg_annual_price.drop([1995,1996,1997,1998,1999,2021],axis=1)

# Step 3: Make 4 lists, each containing 8 borough names, to be used for plotting later. 

list_1_of_boroughs=list(avg_annual_price.iloc[0:8]["London_Borough"])
list_2_of_boroughs=list(avg_annual_price.iloc[8:16]["London_Borough"])
list_3_of_boroughs=list(avg_annual_price.iloc[16:24]["London_Borough"])
list_4_of_boroughs=list(avg_annual_price.iloc[24:]["London_Borough"])

# Step 4: Melt the pivot table. Set index to Borough, Year and then sort 

avg_annual_price_tall = avg_annual_price.melt(id_vars='London_Borough',var_name='year',value_name='average_annual_price')
avg_annual_price_tall = avg_annual_price_tall.set_index(["London_Borough","year"]).sort_index()

# Step 5: Calculate an annual_price_index for each Borough, with the index set to 100 for the year 2000.
#         This index reflects the increase in the each borough's property prices as compared to prices in year 2000.

for index, row in avg_annual_price_tall.iterrows():
  
    avg_annual_price_tall.loc[index, "annual_price_indexed_to_2000"] = \
    (avg_annual_price_tall.loc[index, "average_annual_price"] / \
     avg_annual_price_tall.loc[(index[0],2000.0), "average_annual_price"])*100

avg_annual_price_tall.head()   
```

```python
# Create a pivotable of the Annual Price Index from 2000 to 2020, for the 32 boroughs

avg_annual_price_tall = avg_annual_price_tall.reset_index()
annual_price_index = avg_annual_price_tall.pivot_table(values="annual_price_indexed_to_2000",index="year",columns="London_Borough")

annual_price_index
```

```python
# Plot a sorted horizontal bar chart of the 2020 Price Index 

annual_price_index.loc[2020].sort_values().plot(kind='barh', figsize=(8, 10), width=0.85)

for i,v in enumerate(annual_price_index.loc[2020].sort_values()):
    plt.text(v + 1, i-0.25, str(round(v)), color='blue', fontweight='bold')

plt.xlabel('2020 Price Index')
plt.title('London Boroughs 2020 Average Prices indexed to  year 2000\n\n Index for year 2000 set at 100 for each Borough')
plt.show()


```
